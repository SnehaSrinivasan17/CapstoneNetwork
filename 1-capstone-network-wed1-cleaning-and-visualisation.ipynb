**Infinite values**

* only Flow Byt/s and Flow Pkt/s have infinite values
* i think we will have to replace all the rows which contain infinite values as, i noticed that every row which has the infinity value has different values in other columns in the row
* some may be duplicate and well have to remove those under duplicate function
* Flow Pkt/s is there only in infiltration and sql injection and this dataset is only for bruteforce, so the values may not be necessary and we can drop them

```python
ds = wed1.isin([np.inf, -np.inf])
print("Infinite values before:" ,ds)
```

```python
wed1_data = wed1.iloc[:, 3:79]
wed1_inf_count = np.isinf(wed1_data.iloc[1:]).sum().sum()
print("Count of Infinite Values before:", wed1_inf_count)
```

```python
wed1_inf_values = np.isinf(wed1_data.iloc[1:]).sum()
print("Infinite values count before:", wed1_inf_values.to_string())
```

```python
#to remove infinity values 
wed1 = wed1.replace(["Infinity", "infinity"], np.inf)
wed1 = wed1.replace([np.inf, -np.inf], np.nan)
wed1.replace([np.inf, -np.inf], np.nan, inplace=True)
wed1.dropna(inplace=True)
  
```

```python
ds = wed1.isin([np.inf, -np.inf])
print("Infinite values after:" ,ds)
```

```python
wed1_data = wed1.iloc[:, 3:79]
wed1_inf_count = np.isinf(wed1_data.iloc[1:]).sum().sum()
print("Count of Infinite Values after:", wed1_inf_count)
```

```python
wed1_inf_values = np.isinf(wed1_data.iloc[1:]).sum()
print("Infinite values count after:", wed1_inf_values.to_string())
```
